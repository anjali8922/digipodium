# simple visualization(univariate)

   - histogram -> `.hist()`
   - box plot -> `.plot(kind='box)`
   - area plot -> `.plot(kind='area)`
- For categoricl data
  -  bar plot ->`.plot(kind='bar')`
  - pie chart ->`.plot(kind='pie')`
